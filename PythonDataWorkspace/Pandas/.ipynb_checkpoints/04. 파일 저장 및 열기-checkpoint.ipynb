# 4. 파일 저장 및 열기
DataFrame 객체를 excel, csv, txt 등 형태의 파일로 저장 및 열기

```python
import pandas as pd
data = {
    '이름':['채치수','정대만','송태섭','서태웅','강백호','변덕규','황태산','윤대협'],
    '학교':['북산고','북산고','북산고','북산고','북산고','능남고','능남고','능남고'],
    '키': [197,184,168,187,188,202,188,190],
    '국어': [90, 40, 80, 40, 15, 80, 55, 100],
    '영어': [85, 35, 75, 60, 20, 100, 65, 85],
    '수학': [100, 50, 70, 70, 10, 95, 45, 85],
    'SW': ['Python', 'Java', 'JS', '','','C', 'PYTHON', 'C#']
}
data
df = pd.DataFrame(data, index=['1번', '2번','3번', '4번', '5번', '6번', '7번', '8번'])
df
```

## 저장하기
### CSV 파일로 저장

```python
df.to_csv('score.csv') # CSV 로 저장함
df.to_csv('score.csv', encoding='utf-8-sig', index=False)
```

## 텍스트 파일로 저장

```python
df.to_csv('score.txt', sep='\t') # tab 으로 구분된 텍스트 파일
```

## CSV 파일 열기

```python
df = pd.read_csv('score.csv')
df
```

```python
df = pd.read_csv('score.csv', skiprows=1)
df
```

```python
df = pd.read_csv('score.csv', skiprows=[1,3,5]) # 1,3,5 row 는 제외 (row 0 부터 시작)
df
```
